```python
#Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

import warnings
warnings.filterwarnings('ignore')
```

```python
#loading the dataset
ds = pd.ExcelFile('Climate_Data_RF_&_Temp_2012_to_2018.xlsx')
```

```python
#Loading the MONTHLY RAINFALL TOTALS (mm) data
Rain_df = pd.read_excel(ds,sheet_name='Rainfall',header=1)
Rain_df
```

```python
#Loading the Mean Monthly Maximum Temperature (0C) data
Max_temp_df = pd.read_excel(ds,sheet_name='Temp_Max',header=3)
Max_temp_df
```

```python
#Loading the Mean Monthly Minimum Temperature (0C) data

Min_temp_df = pd.read_excel(ds,sheet_name='Temp_Min',header=3)
Min_temp_df.head()
```

```python
dataset = {
    'Rain':Rain_df, 'minTemp':Min_temp_df,'maxTemp':Max_temp_df
}
for key, data in dataset.items():
    print(f"Data {key}:{data}")
```

```python
#Null value imputation for the columns
for col in Rain_df.columns:
    if Rain_df[col].isnull().sum()>0:
        val1 = Rain_df[col].mean()
        Rain_df[col] = Rain_df[col].fillna(val1)
for col in Min_temp_df.columns:
    if Min_temp_df[col].isnull().sum()>0:
        val2 = Min_temp_df[col].mean()
        Min_temp_df[col] = Min_temp_df[col].fillna(val2)
for col in Max_temp_df.columns:
    if Max_temp_df[col].isnull().sum()>0:
        val3 = Max_temp_df[col].mean()
        Max_temp_df[col] = Max_temp_df[col].fillna(val2)
        
print(f"Rainfall: {Rain_df.isnull().sum().sum()}")
print(f"Min Temp: {Min_temp_df.isnull().sum().sum()}")
print(f"Max Temp: {Max_temp_df.isnull().sum().sum()}")
```
